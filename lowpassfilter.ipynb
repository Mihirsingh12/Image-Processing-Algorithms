```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
img=mpimg.imread('einstein2.jpg')
x=img[:,:,0]
plt.imshow(x,cmap="gray")
plt.show()
y=x
y.setflags(write=1)
```

```python
t1=list(np.shape(x))
t1
```

```python
for i in range(t1[0]-2):
    for j in range(t1[1]-2):
        s=y[i:(i+3),j:(j+3)]
        mean=(np.sum(s)/9)
        if mean-int(mean)>=0.5:
            mean=int(mean)+1
        else:
            mean=int(mean)
        y[i+1,j+1]=mean
plt.imshow(y,cmap="gray")
plt.show()
```
