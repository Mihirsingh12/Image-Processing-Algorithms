```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
def d2b(n):
    return list(map(int,bin(n).replace("0b","").zfill(8)))
def rep(m,n):
    i=len(m)-1
    j=0
    while j<n:
        m[i]=0
        j=j+1
        i=i-1
    return m
fig,ax=plt.subplots(2,4,figsize=(20,10))
fig.subplots_adjust(wspace=0,hspace=0)
img=mpimg.imread('einstein.jpg')
x=img[:,:,0]
ax[0,0].imshow(x,cmap="gray")
ax[0,0].axis('off')
y=np.zeros(np.shape(x))
t1=list(np.shape(x))
temp=[]
m=1
a=0
b=1
while m<8:
    for i in range(t1[0]):
        for j in range(t1[1]):
            temp=d2b(x[i][j])
            temp1=rep(temp,m)
            temp2=int(''.join(map(str,temp1)),2)
            y[i][j]=temp2
    if m==4:
        a=1
        b=0
    ax[a,b].imshow(y,cmap="gray")
    ax[a,b].axis('off')
    b=b+1
    m=m+1
```
