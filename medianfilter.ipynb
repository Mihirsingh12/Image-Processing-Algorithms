```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
img=mpimg.imread('mf3.jpg')
x=img[:,:,0]
y=np.zeros(np.shape(x))
t1=list(np.shape(x))
y=x
y.setflags(write=1)
fig,ax=plt.subplots(1,3,figsize=(20,10))
ax[0].imshow(x,cmap="gray")
ax[0].axis('off')
#print(x.flags)
#print(y.flags)
#plt.imshow(x,cmap="gray")
#plt.show()
for i in range(t1[0]-2):
    for j in range(t1[1]-2):
        s=y[i:(i+3),j:(j+3)]
        med=np.median(s)
        y[i+1][j+1]=int(med)
ax[1].imshow(y,cmap="gray")
ax[1].axis('off')
#plt.show()
for i in range(t1[0]-4):
    for j in range(t1[1]-4):
        s=y[i:(i+5),j:(j+5)]
        med=np.median(s)
        y[i+2][j+2]=int(med)
ax[2].imshow(y,cmap="gray")
ax[2].axis('off')
#plt.show()
```
