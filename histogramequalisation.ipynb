```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
img=mpimg.imread('einstein2.jpg')
x=img[:,:,0]
plt.imshow(x,cmap="gray")
plt.show()
y=np.zeros(np.shape(x))
sum1=0
t2=list(np.shape(x))
t1=list()
totalpixels=t2[0]*t2[1]
for i in range(t2[0]):
    for j in range(t2[1]):
        if x[i][j] not in t1:
            t1.append(x[i][j])
t1.sort()
l=len(t1)
for i in range(l):
    sum1=sum1+(x==t1[i]).sum()
    d=(sum1/totalpixels)*255
    if d-int(d)>=0.5:
        d=int(d)+1
    else:
        d=int(d)
    y[x==t1[i]]=d
plt.imshow(y,cmap="gray")
plt.show()
#plt.hist(y)
```
