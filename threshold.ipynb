```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
img=mpimg.imread('heart.jpg')
x=img[:,:,0]
plt.imshow(x,cmap="gray")
plt.show()
y=np.zeros(np.shape(x))
t2=list(np.shape(x))
av=np.average(x)
for i in range(t2[0]):
    for j in range(t2[1]):
        if x[i][j]<=av/2:
            y[i][j]=255
        else :
            y[i][j]=0
plt.imshow(y,cmap="gray")
plt.show()
img=mpimg.imread('threshold.jpg')
x=img[:,:,0]
plt.imshow(x,cmap="gray")
plt.show()
av=np.average(x)
y=np.zeros(np.shape(x))
t2=list(np.shape(x))
for i in range(t2[0]):
    for j in range(t2[1]):
        if x[i][j]<50:
            y[i][j]=x[i][j]
        elif x[i][j]>=av and x[i][j]<=(2*av):
            y[i][j]=8*x[i][j]
        else :
            y[i][j]=8*x[i][j]
plt.imshow(y,cmap="gray")
plt.show()
```
